# Reading in the csv file and putting it in a dataframe

```python
import pandas as pd

df = pd.read_csv('Tweets.csv')
df
```

# Making every column with words lowercase

```python
df['negativereason'] = df['negativereason'].str.lower()
df['airline'] = df['airline'].str.lower()
df['name'] = df['name'].str.lower()
df['text'] = df['text'].str.lower()
df['tweet_location'] = df['tweet_location'].str.lower()
df['user_timezone'] = df['user_timezone'].str.lower()
```

```python
df
```

Printing the dataframe to make sure the values became lowercase

# Removing stop words and punctuation from the tweets

```python
import nltk
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df['text']
```

# Finding and filling NA values

```python
df.isna().sum()
```

```python
df = df.fillna(df.mode().iloc[0])
```

Filled the NA vaules with most common value for each column

```python
df.isna().sum()
```

Checking to make sure the NA values were filled

# Removing puncuation from the tweets and tokenizing the words

```python
df['text'] = df['text'].str.replace('[^\w\s]','')
df['text']
```

Removing the puncuation from the tweets 

# Using tenssoflow tokenizer to tokenize the text and put it in a new column

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
t = Tokenizer()
t.fit_on_texts(df['text'])
text_tokenized = t.texts_to_sequences(df['text'])

text_tokenized_pad = tf.keras.preprocessing.sequence.pad_sequences(text_tokenized, maxlen=2, truncating='pre')

df2 = pd.concat([df, pd.DataFrame(text_tokenized_pad, columns=['text','tokenized_text'])], axis=1)
df2
```

Created a second dataframe to store the values of the original dataframe and the new tokenized text

```python
df2.drop(['text'], axis =1, inplace = True)
df2
```

Droping the text column since it is not tokenized

# Label encoding the airlien_sentiment column

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df2['airline_sentiment'] = le.fit_transform(df2['airline_sentiment'])
df2
```

Assigning 1 to netural, 2 to positive, and 0 to negative

# Splitting the dataset in to train and test

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df,test_size=0.2,random_state=42)
```

# Setting up and scoring attaboost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification
train, test = make_classification(n_samples=100, n_features=4,n_informative=2, n_redundant=0,random_state=0, shuffle=False)
clf = AdaBoostClassifier(n_estimators=100, random_state=0).fit(train, test)
AdaBoostClassifier(n_estimators=100, random_state=0)
clf.score(train, test)
```

The score for attaboost is 1.0

# Setting up and running bagging

```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification
train,test = make_classification(n_samples=100, n_features=4,n_informative=2, n_redundant=0,random_state=0, shuffle=False)
clfbagging = BaggingClassifier(base_estimator=SVC(),n_estimators=2, random_state=0).fit(train, test)
clfbagging.score(train, test)
```

I made sure to give each of these the same number of n_samples, n_informative, n_redundant, random_state, and Shuffle. Bagging scored .89

# Setting up and running random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
train, test = make_classification(n_samples=100, n_features=4,n_informative=2, n_redundant=0,random_state=0, shuffle=False)
clfrf = RandomForestClassifier(max_depth=2, random_state=0).fit(train, test)
clfrf.score(train,test)
```

Lastly random forest scored .93

Of the 3 Attaboost did the best and had the highest accuracy score and bagging had the lowest. I think attaboost did the best because it generates more than one weak learner and then eventally outputs the one that is most accurate. While random forest and bagging use randomness
